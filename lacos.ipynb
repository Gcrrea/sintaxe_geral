```python
import pandas as pd
```

```python
dados = pd.read_excel(r"C:\estudo_datalake\Sample - Superstore.xls")
dicts = dados.to_dict('records')
```

```python
#Iterando com Laço for. Caso nao encontre a chave ele gera exceção sem quebrar pois estou usando dict.get('Quantity',0)
for dict in dicts:
    new_dicts = {
        "ID":dict.get('Row ID',0),
        "Order":dict.get('Order ID',0),
        "data":dict.get('Order Date',0),
        "Valor":dict.get('Sales',0),
        "Quantidade":dict.get('Quantity',0)
    }
    print(new_dicts)
    
```

```python
#Gerando nova lista de dicionarios

lista_dicionarios = []

for dict in dicts:
    new_dicts = {
        "ID":dict.get('Row ID'),
        "Order":dict.get('Order ID'),
        "data":dict.get('Order Date'),
        "Valor":dict.get('Sales'),
        "Quantidade":dict.get('Quantity')
    }
    print(new_dicts)
    lista_dicionarios.append(new_dicts)

print(lista_dicionarios)
```

```python
#Usando laço while para iterar

cont = len(dicts)
i = 0

while i <= cont:
    try:
        print(dicts[i])
    except:
        pass 
    
    i+=1
```

```python
#Laço while aplicando condições

cont = len(dicts)
i = 0

while i <= cont:
    try:
        print(dicts[i])
        if dicts[i]['Row ID'] > 9993:
            print('acho', dicts[i]['Row ID'])
        else:
            print("Nao é",dicts[i]['Row ID'])
    except:
        pass 
    
    i+=1
```
