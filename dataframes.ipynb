```python
# ascending=False Ordena do maior para o menor
# ascending=False Ordena do menor para o maior

ordenando = dados.sort_values('Order Date', ascending=False)
ordenando.head(2)
ordenando[['Customer Name','Order Date']]
```

```python
#faz contagem dos valores da coluna Regiao
dados['Region'].value_counts()
```

```python
#faz agrupamento pelas colunas "Region","Order Date","Customer Name" somando  valor vendido de cada item
dados.groupby(["Region","Order Date","Customer Name"])["Sales"].sum().reset_index()
```

```python
#Iterando em dataframe. No final gerou exceção "erro" por causa de coluna invalida porem eu tratei o erro com try/except
try:
    for i, row in dados.iterrows():
        print(row[i])
except Exception as e:
    print(e)
```
